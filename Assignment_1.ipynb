```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files
uploaded = files.upload()

sales = pd.read_csv("Housing.csv")
sales.head()
```

```python
plt.figure(figsize=(10,8))

corr = sales.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(8,6))
avg_price_by_furnishing = sales.groupby('furnishingstatus')['price'].mean().reset_index()
sns.barplot(x='furnishingstatus', y='price', data=avg_price_by_furnishing)
plt.title('Average Price by Furnishing Status')
plt.xlabel('Furnishing Status')
plt.ylabel('Average Price (in Rs)')
plt.grid(True)
plt.show()
```

```python

plt.figure(figsize=(8,6))
sns.boxplot(x='bedrooms', y='price', data=sales)
plt.title('Price Distribution by Number of Bedrooms')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Price (in Rs)')
plt.grid(True)
plt.show()
```

```python

plt.figure(figsize=(8,6))
plt.hist(sales['bedrooms'], bins=5, color='green', edgecolor='black')
plt.title('Distribution of Number of Bedrooms')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,6))
plt.scatter(sales['area'], sales['price'], color='b') # Changed df to sales
plt.title('Price vs Area')
plt.xlabel('Area (in sqft)')
plt.ylabel('Price (in Rs)')
plt.grid(True)
plt.show()
```
